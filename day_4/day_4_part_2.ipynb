```python
from collections import defaultdict
def read_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()

# content = read_file('samples/sample_2_18.txt')
content = read_file('./challange.txt')

def letter_to_coordinates(content):
    letter_coordinates = defaultdict(list)
    for i, line in enumerate(content.split('\n')):
        for j, letter in enumerate(line):
            letter_coordinates[letter].append((i, j))
    return letter_coordinates

def coordinate_to_letter(coordinates):
    return {coordinate: letter for letter, coordinates in coordinates.items() for coordinate in coordinates}


letter_coordinates = letter_to_coordinates(content)
coordinates_to_letter = coordinate_to_letter(letter_coordinates)

bounds = (max(coordinates_to_letter.keys(), key=lambda x: x[0])[0] + 1, max(coordinates_to_letter.keys(), key=lambda x: x[1])[1] + 1)
bounds
```

```python
def get_corner_coordinates(coordinate):
    x, y = coordinate
    corners = []
    for i in (-1,1):
        for j in (-1,1):
            corners.append((x+i, y+j))
    return corners

def get_opposite_corners(coordinate):
    x, y = coordinate
    return [(x-1, y-1), (x+1, y+1)], [(x-1, y+1), (x+1, y-1)]


def is_cross_mas(coordinate, bounds):
    # If on boundary skip
    if coordinate[0] == 0 or coordinate[0] == bounds[0] - 1 or coordinate[1] == 0 or coordinate[1] == bounds[1] - 1:
        return False
    
    corners = get_corner_coordinates(coordinate)
    # If any corner is not M or S, return False
    for corner in corners:
        if coordinates_to_letter[corner] not in ['M', 'S']:
            return False
    
    # If opposite corners are equal, return False
    opposite_corners = get_opposite_corners(coordinate)
    for corner in opposite_corners:
        if coordinates_to_letter[corner[0]] == coordinates_to_letter[corner[1]]:
            return False
    return True

words = [("A", coordinates) for coordinates in letter_coordinates["A"]]

count = 0
words_found = []
for (letter, coordinate) in words:
    success = is_cross_mas(coordinate, bounds)
    if success:
        count += 1


count
```
